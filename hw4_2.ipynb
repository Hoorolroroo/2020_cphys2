```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
iris = datasets.load_iris()
X=iris["data"]
y=iris["target"]
```

## Support Vector Machine

1.SVM 기법을 이용하기 전에, 

'sepal length (cm)',
'sepal width (cm)',
'petal length (cm)',
'petal width (cm)'중 어떤 요인이 종의 분류의 가장 큰 요인인지 알아보자.

이번에는 피팅이 아니라, 순수하게 나열을 통한 직관을 해보겠습니다.

(1) sepal length

```python
svm=SVC(kernel='rbf', gamma='auto')
svm.fit(X, y)
plt.scatter(X[:,0], y, c=y)
```

(2) sepal width

```python
plt.scatter(X[:,1], y, c=y)
```

(3) petal length

```python
plt.scatter(X[:,2], y, c=y)
```

위의 두 데이터보다 확연히 선형적인 차이를 보이고 있습니다.

(4) petal width

```python
plt.scatter(X[:,3], y, c=y)
```

다른 데이터와 달리 비교적 확연한 구분선이 보이는 꽃잎의 길이와 너비를 X축 Y축으로 하는 그래프에, SVM 기법으로 새로 피팅해 보겠습니다.

```python
plt.figure(figsize=(10,5))
plt.scatter(X[:,2], X[:,3], c=svm.predict(X))
plt.xlabel("Petal length", fontsize='x-large')
plt.ylabel("Petal width", fontsize='x-large')
```

```python
svm.score(X, y)
```

위의 그래프와 같이 꽃잎의 길이와 너비를 축으로 하는 그래프에서 높은 스코어와 선형적인 데이터를 얻었습니다.

즉 Iris의 세 종의 분류하는 기준은 꽃잎의 길이와 너비라는 결론을 내릴 수 있습니다.

## ※ 최종 결론


앞서 Logistic Regression 피팅과도 유사한 결론이 나왔으며, 여기서는 versicolor의 분석도 가능하게 나왔습니다.

1. Iris 종 중에서 꽃잎의 길이와 너비가 가장 작은 종은 **Setosa**이다. (꽃잎의 길이가 1.0 ~ 2.0 사이, 너비는 0.2~ 0.5 사이이다.단위(cm))


2. 그 중간은 **versicolor**이다.(꽃잎의 길이가 3.0 ~ 5.0 사이, 너비는 1.0 ~ 1.5 사이이다.)


3. 꽃잎의 길이와 너비가 가장 큰 종은 **versinica**이다.(꽃잎의 길이는 대략 5.0 ~ 7.0사이,너비는 1.5 ~ 2.5 사이이다. )
